```python
# Foundations of Machine Learning
# IT 582
# Student-ID : 202411012
# Name : Manish Manojkumar Prajapati

# Problem 2
# Plot theta vs. L(theta), where L(theta) = theta_1^2 + theta_2^2
# theta_1 and theta_2 vary from -10 to +10 with a step size of 0.1.
# Locate the minimum value of L(theta_1, theta_2) with corresponding theta_1,
# theta_2 values from the plot.


# Importing required libraries

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Creating loss function

def lossFunction(theta_1, theta_2):
  return theta_1**2 + theta_2**2


# Defining the step size

step = 0.1


# Generating values for THETA_1 and THETA_2

theta_1 = np.arange(-10,10.1,step)
theta_2 = np.arange(-10,10.1,step)


# Creating Mesh Grid
# MeshGrid allows us to compute the function values for each combination of
# THETA_1 and THETA_2 in a 2-D grid

theta_1_mesh, theta_2_mesh = np.meshgrid(theta_1, theta_2)


# Calculating the loss function values for all combinations of THETA_1 and
# THETA_2 values

loss_function_values = lossFunction(theta_1_mesh, theta_2_mesh)



# figure() function of pyplot is used to create a new figure
# This is used here in order to increase the size of the figure

plt.figure(figsize=(8,6))


# Contour plot allows to visualize 3-D data into a two-dimensional plot

contour = plt.contourf(theta_1_mesh, theta_2_mesh, loss_function_values, cmap="viridis")
plt.colorbar(contour, label="L(THETA) = THETA1^2 + THETA2^2")


# Labeling axis' of graph

plt.xlabel('Theta_1')
plt.ylabel('Theta_2')


# Entitling graph

plt.title("Contour Plot of L(THETA1, THETA2) = THETA^2 + THETA^2")
plt.grid(True)


# Finding minimum Loss function value of L(theta1, theta2)
min_L_value = np.min(loss_function_values)


# np.unravel_index is used to convert a flat (one-dimesional) index into a
# tuple of coordinate indices for any array of a specified shape

min_theta_1_index, min_theta_2_index = np.unravel_index(np.argmin(loss_function_values), loss_function_values.shape)


# Fetching the minimum theta_1 and theta_2 values

min_theta_1_value = theta_1[min_theta_1_index]
min_theta_2_value = theta_2[min_theta_2_index]

plt.scatter(min_theta_1_value, min_theta_2_value, color = "red", marker="X", label=f"Min L(theta1, theta2) = {min_L_value}\n at theta1 = {min_theta_1_value}, theta = {min_theta_2_value}")
plt.legend()

plt.show()


# Making a corresponding 3-D graph

plt_2 = plt


# figure() function is used to create a figure of specific size
# This is used for enlarging the graph size

figure = plt_2.figure(figsize=(10,7))

# add_subplot method allows us to add an individual subplot to a figure
# This method is a part of the 'Figure' class in Matplotlib, and its used to
# create a grid of subplots within a single figure

ax = figure.add_subplot(111, projection='3d')
surface = ax.plot_surface(theta_1_mesh, theta_2_mesh, loss_function_values, cmap="viridis")


# Adding labels and title to 3D graph plot

ax.set_xlabel('THETA_1')
ax.set_ylabel('THETA_2')
ax.set_zlabel('L(THETA_1, THETA_2)')
ax.set_title('Surface plot of L(THETA_1, THETA_2) = THETA_1^2 + THETA_2^2')


# colorbar() adds a colorbar to the plot, which is visual representation of
# the range of values in a colormap

figure.colorbar(surface, ax = ax, shrink=0.5, aspect=5)
```

```python
# Foundations of Machine Learning
# IT 582
# Student-ID : 202411012
# Name : Manish Manojkumar Prajapati

# Problem 3
# a.) Plot for L(theta) = SUM from i = 1 to m (y(i) - (theta_0 + theta_1.x(i)))^2 / m,
#     where m is the number of input examples and x(i), y(i) are the values taken from
#     the given dataset. Obtain the minimum value of L(theta) with corresponding
#     theta_0, theta_1 values from the plot.
#


# Importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Here the r infront of string declares the string as raw string and signals the reader to read address string considering it raw

url = r'https://raw.githubusercontent.com/ManishPraa24/Linear-Regression/main/Data.csv'


# Fetching the data.csv file into data variable

data = pd.read_csv(url)


# Knowing the data through data.describe()

print(data.describe())


# Defining the Loss function

def loss_function(theta_0, theta_1, x, y):  # These are the parameters, x is a feature vector and y is the corresponding value
    
    m = len(x)

    total_error = np.sum((y- (theta_0 + (theta_1 * x)))**2)
    return total_error/(2*m)


# Fetching feature vector values

X = data['x'].values


# Fetching the target vector values

Y = data['y'].values

print()
# print("Y = ", Y)


# Initializing theta_0 and theta_1

theta_0 = np.linspace(-100, 100, len(X)) # Changing it to -1 to 1
theta_1 = np.linspace(-0.8, 0.8, len(Y)) # Changing it to -1 to 1


# Calculating loss_function values for all combinations of theta_0 and theta_1

loss_function_values = list()

for first_theta in theta_0:
    for second_theta in theta_1:
        loss = loss_function(first_theta, second_theta, X, Y)
        loss_function_values.append((first_theta, second_theta, loss))


# Converting loss_function_values to numpy array for efficient accessing of elements alongwith applying specific members

loss_values = np.array(loss_function_values)


# Extracting theta_0, theta_1, and loss for plotting
# Because, extracting from numpy's array with boost the plotting

theta_0_plot = loss_values[:,0]
theta_1_plot = loss_values[:,1]
loss_plot = loss_values[:,2]


# Extracting the minimum value of loss with its corresponding theta_0 and theta_1 value

min_index = np.argmin(loss_plot)
loss_optimal = loss_plot[min_index]
theta_0_optimal = theta_0_plot[min_index]
theta_1_optimal = theta_1_plot[min_index]

plt.figure(figsize=(10,6))


sc = plt.scatter(theta_0_plot, theta_1_plot, c=loss_plot, cmap="viridis") 


# Marking the theta_0_optimal and theta_1_optimal point on the graph

plt.scatter(theta_0_optimal, theta_1_optimal, color = "red", marker="X", label=f"Min L(theta0, theta1) = {loss_optimal}\n at theta0 = {theta_0_optimal}, theta1 = {theta_1_optimal}")
plt.legend()
plt.xlabel('theta_0')
plt.ylabel('theta_1')
plt.title('Scatter plot of the Loss Function')
plt.colorbar(sc, label='Loss')
plt.show()

# Making a corresponding 3-D graph

plt_2 = plt


# figure() function is used to create a figure of specific size
# This is used for enlarging the graph size

figure = plt_2.figure(figsize=(12,8))

# add_subplot method allows us to add an individual subplot to a figure
# This method is a part of the 'Figure' class in Matplotlib, and its used to
# create a grid of subplots within a single figure

ax = figure.add_subplot(111, projection='3d')


# Used scatter here in-order to plot individial data points in a 3-D space
# Here, loss_plot will give a z-coordinate to the plane of theta_0 and theta_1 plots
surface = ax.scatter(theta_0_plot, theta_1_plot, loss_plot, c=loss_plot, cmap="viridis")


# Adding labels and title to 3D graph plot

ax.set_xlabel('THETA_0')
ax.set_ylabel('THETA_1')
ax.set_zlabel('L(THETA_0, THETA_1)')
ax.set_title('Surface plot of L(THETA_0, THETA_1) = THETA_0^2 + THETA_1^2')


# colorbar() adds a colorbar to the plot, which is visual representation of
# the range of values in a colormap

figure.colorbar(surface, ax = ax, shrink=0.5, aspect=5)

print()
print(f"The minimum loss value obtained is {loss_optimal}.")
print(f"The corresponding theta_0 value is {theta_0_optimal}.")
print(f"The corresponding theta_1 value is {theta_1_optimal}.")
```

<h3>Why can scikit-learn implementation still correctly solve this regression problem? </h3>

    Scikit-learn do not rely directly on the normal equation method to solve linear regression problem.
    </li>
